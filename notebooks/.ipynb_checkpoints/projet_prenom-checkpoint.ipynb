```python
def prop_prenoms_region(df, liste, annee, prenom):
    """ Renvoie la proportion d'un certain prénom donnée à la naissance dans une région par rapport à la France 
    (df = DataFrame prénoms, liste = liste de départements de la région étudié, annee = annee à étudier), prenom = prenom à étudier"""
    prop = df_dpt[(df_dpt['dpt'].isin(liste)) & (df_dpt['année'] == annee) & (df_dpt['preusuel'] == prenom)].groupby('année').sum()['nombre'].iloc[0]/ df_dpt[(df_dpt['année'] == annee) & (df_dpt['preusuel'] == prenom)].groupby('année').sum()['nombre'].iloc[0]
    return prop
```

```python
############ Etudier la présence de certains prénoms dans toute la France et le cantonnement à une région de certains
########### Pourcentage de Erwan et Enora nés en Bretagne en 2017 et de Maelys nés en Occitanie
liste_dpt_occitanie = [65, 32, 82, 46, 12, 48, 30, 34, 11, 66, 9, 11, 31, 81]
print("La proportion de Erwan nés en Bretagne par rapport à ceux nés dans toute la france est de", prop_prenoms_region(df_dpt, liste_dpt_bretagne, 2017, 'ERWAN'))
print("La proportion d'Enora nés en Bretagne par rapport à ceux nés dans toute la france est de", prop_prenoms_region(df_dpt, liste_dpt_bretagne, 2017, 'ENORA'))
print("La proportion de Maelys nés en Occitanie par rapport à ceux nés dans toute la france est de", prop_prenoms_region(df_dpt, liste_dpt_occitanie, 2017, 'MAELYS'))
```

```python
##################### Evolution de la proportion des prénoms
def serie_prop_prenom(df, prenom):
    df_prenom = df[df['preusuel'] == prenom]
    serie_prop = df_prenom.groupby('année').sum()['nombre']/df.groupby('année').sum()['nombre']
    serie_prop = serie_prop.dropna(axis=0)
    return serie_prop
```

```python
########################## prénoms limités à une région: Armel (Bretagne), Maite (basque), Giulia (Italie)
def nb_prenoms_par_dpt(df, prenom):
    df_prenom = df[df['preusuel'] == prenom] 
    serie_prenom_dpt = df_prenom.groupby('dpt').sum()['nombre']
    return serie_prenom_dpt
```

```python
def serie_prop_prenom_region(df, prenom, liste_dpt):
    df_prenom = df[df['preusuel'] == prenom]
    serie_prop_region_prenom = df_prenom[df_prenom['dpt'].isin(liste_dpt)].groupby('année').sum()['nombre'] / df_prenom.groupby('année').sum()['nombre']
    return serie_prop_region_prenom
```

```python
liste_dpt_basque = [64, 65, 31, 9, 66, 11, 32, 40]
liste_dpt_italie = [4, 5, 6, 73, 38, 74, 13, 83]
serie_prop_region_armel = serie_prop_prenom_region(df_dpt, 'ARMEL', liste_dpt_bretagne)
serie_prop_region_maite = serie_prop_prenom_region(df_dpt, 'MAITE', liste_dpt_basque)
serie_prop_region_armel = serie_prop_prenom_region(df_dpt, 'GIULIA', liste_dpt_italie)
serie_armel_dpt = nb_prenoms_par_dpt(df_dpt, 'ARMEL')
serie_maite_dpt = nb_prenoms_par_dpt(df_dpt, 'MAITE')
serie_giulia_dpt = nb_prenoms_par_dpt(df_dpt, 'GIULIA')

nrows2 = 3
ncols2 = 2
fig2, axes2 = plt.subplots(nrows2, ncols2)
serie_armel_dpt.plot.bar(ax = axes2[0,0])
serie_maite_dpt.plot.bar(ax = axes2[1,0])
serie_giulia_dpt.plot.bar(ax = axes2[2,0])
serie_prop_region_armel.plot(ax = axes2[0,1])
serie_prop_region_maite.plot(ax = axes2[1,1])
serie_prop_region_armel.plot(ax = axes2[2,1])
fig2.set_figheight(20)
fig2.set_figwidth(25)

liste_prenoms2 = ['Armel', 'Maite', 'Giulia']
liste_regions = [' en Bretagne', ' au Pays Basque', ' à la Frontière italienne']
for c in range(ncols2):
    ind_pre2 = 0
    for r in range(nrows2):
        if(c == 0):
            axes2[r,c].set_xlabel('Numéro de département')
            axes2[r,c].set_ylabel('Nombre de ' +  liste_prenoms2[ind_pre2])
            axes2[r,c].set_title('Nombre de ' + liste_prenoms2[ind_pre2] + ' en fonction du département')
        else:
            axes2[r,c].set_xlabel('Année de naissance')
            axes2[r,c].set_ylabel('Proportion de ' +  liste_prenoms2[ind_pre2] + liste_regions[ind_pre2])
            axes2[r,c].set_title('Proportion  ' + liste_prenoms2[ind_pre2] + liste_regions[ind_pre2] + ' en fonction du temps')
        ind_pre2 += 1
```

```python
def evol_nb_prenom(df, prenom):
    serie_nb_prenom_annee = df[df['preusuel'] == prenom].groupby('année').sum()['nombre']
    return serie_nb_prenom_annee
```

```python
######################### Evolution de la proportion des noms français en fonction du temps
def serie_prop_langage(df, langage):
    df_langage = df[df['langage'].str.contains(langage)]
    serie_prop = df_langage.groupby('année').sum()['nombre']/df.groupby('année').sum()['nombre']
    serie_prop = serie_prop.dropna(axis=0)
    return serie_prop
```

```python
def df_prop_langages(df, list_langages):
    list_series_langage = []
    for ind, langage in enumerate(list_langages):
        list_series_langage.append(df[df['langage'].str.contains(langage)])
        list_series_langage[ind] = list_series_langage[ind].groupby('année').sum()['nombre']/df.groupby('année').sum()['nombre']
        list_series_langage[ind] = list_series_langage[ind].dropna(axis=0)
    df_prop_langage = list_series_langage[0].to_frame()
    for ind_serie in range(1, len(list_series_langage)):
        df_prop_langage = df_prop_langage.merge(list_series_langage[ind_serie], left_index = True, right_index = True, suffixes=('',str(ind_serie)))
    df_prop_langage.columns = list_langages
    return df_prop_langage
```

```python
list_langages = ['french', 'biblical', 'english', 'portuguese', 'italian', 'german', 'arabic', 'scandinavian', 'breton', 'catalan', 'spanish']
df_langages = df_prop_langages(df_merged_origins, list_langages)
df_langages.sample(10)
```

```python
####################### Carte du nombre et de la proportion des naissances en France par départements sur une année
def serie_nb_par_dpt(df, annee):
    serie = df[df['année'] == annee].groupby('dpt').sum()['nombre']
    return serie
```

```python
def serie_prop_par_dpt(df, annee):
    serie = df[df['année'] == annee].groupby('dpt').sum()['nombre']/df[df['année'] == annee].groupby('dpt').sum().sum()['nombre']
    return serie
```
